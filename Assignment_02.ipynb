Exercise 1: Library Management System

Create a Book class with attributes such as title, author, and isbn. Implement a Library class that manages a collection of books. Include methods to add, remove, and search for books by title or author. Store the book details in a text file.

```python
#class Book with members title, author and isbn
class Book:
  def __init__(self, title, author, isbn):
    self.title = title
    self.author = author
    self.isbn = isbn

  def __str__(self):
    return f"{self.title},{self.author},{self.isbn}"

#class Library with multiple functions
class Library:
  def __init__(self):
    self.books = []

  #This function add books to the list and update the file
  def add(self, book):
    self.books.append(book)
    self.save()

  #Display all the books in the list
  def displayAll(self):
    print("All book in library: ")
    for book in self.books:
      print(book)

  #Remove a selected book by isbn from list and update file
  def remove(self, isbn):
    self.books = [book for book in self.books if book.isbn != isbn]
    print(f"\nRemoved book: {book.title}")
    self.save()

  #Search selected book in list by title
  def search_by_title(self, title):
    return [book for book in self.books if title.lower() in book.title.lower()]

  #Search selected book in list by author
  def search_by_author(self, author):
    return [book for book in self.books if author.lower() in book.author.lower()]

  #Save book details in the txt file call library
  def save(self, filename = "library.txt"):
    with open(filename, "w") as file:
      for book in self.books:
        file.write(str(book) + "\n")

  #Load stored data from file and maket it to readable formart
  def load_books(self, filename = "library.txt"):
    print("\nLoad data from file: ")
    with open(filename, "r") as file:
      for line in file:
        parts = line.strip().split(",")
        if len(parts) == 3:
          title, author, isbn = parts
          book = Book(title, author, isbn)
          print(book)


library = Library()

B1 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084")
B2 = Book("1984", "George Orwell", "9780451524935")
B3 = Book("Pride and Prejudice", "Jane Austen", "9780141439518")
B4 = Book("The Catcher in the Rye", "J.D. Salinger", "9780316769488")

library.add(B1)
library.add(B2)
library.add(B3)
library.add(B4)

library.displayAll()

library.remove("9780061120084")

books_by_title = library.search_by_title("1984")
print("\nSearch by title")
for book in books_by_title:
  print(f"{book.title}, {book.author}, {book.isbn}")

books_by_author = library.search_by_author("Jane")
print("\nSearch by author")
for book in books_by_author:
  print(f"Found by author: {book.title}, {book.author}, {book.isbn}")

library.load_books()
```

Exercise 2: Student Grades System

Create a Student class with attributes like name, id, and a list of grades. Implement methods to add
grades, calculate the average, and determine if the student passed (average >= 60). Store student
details in a JSON file.

```python
import json

#Define Student class with attibute name, id and grades
class Student:
  def __init__(self, name, id, grades=[]):
    self.name = name
    self.id = id
    self.grades = grades

  def __str__(self):
    return f"{self.name}, {self.id}, {self.grades}"

  #Adding grades to list
  def add_grade(self, grade):
    self.grades.append(grade)

  #Calculate the average
  def average(self):
    average = sum(self.grades) / len(self.grades)
    if len(self.grades) == 0:
        return 0
    elif average >= 60:
      return f"Passed : {average}"
    else:
      return f"Failed: {average}"

  #Save data in JSON fromat using json file
  def save(self, filename = "Data.json"):
    x = {
      "name" : self.name,
      "id" : self.id,
      "marks" : self.grades,
      "grade" : self.average()
    }

    with open(filename, "w") as file:
      json.dump(x, file, indent=4)

  #Load JSON data
  def load(self, filename):
    with open(filename, "r") as file:
      data = json.load(file)

    print(json.dumps(data, indent=4))


S1 = Student("name", "CT19")

S1.add_grade(100)
S1.add_grade(80)

print(S1)

S1.average()

S1.save()
S1.load("Data.json")
```

Exercise 3: Task Manager

Implement a Task class with attributes title, description, and status. Create a TaskManager class
that uses a queue to manage tasks (add, complete, and display pending tasks). Save the task list to
a CSV file.

```python
from collections import deque

#Implement the claas Task with following attributes
class Task:
  def __init__(self, title, description, status = "Pending"):
    self.title = title
    self.description = description
    self.status = status

#Implement new class TaskManger that can perfrom follwing functions
class TaskManager:
  def __init__(self):
    self.tasks = deque()

  def add(self, title, description):
    self.tasks.append(Task(title, description))
    print(f"Task {title} added successfully")

  def done(self):
    if self.tasks:
      task = self.tasks.popleft()
      task.status = "Completed"
      print(f"\nTask {task.title} has been done")
    else:
      print("\nNo pending tasks")

  def display(self):
    if self.tasks:
      print("\nPending Tasks:")
      for task in self.tasks:
        print(f"Title: {task.title}, Description: {task.description}, Status: {task.status}")
    else:
      print("\nNo pending tasks")

  def save(self, filename = "tasks.csv"):
    with open(filename, "w") as file:
      for task in self.tasks:
        file.write(f"{task.title},{task.description},{task.status}\n")

TM = TaskManager()

TM.add("Write Report", "Draft the quarterly financial report for review.")
TM.add("Team Meeting", "Discuss project updates and assign new tasks.")
TM.add("Client Follow-up", "Reach out to the client for feedback on the latest deliverable.")
TM.add("Code Review", "Review the pull requests from the development team for quality assurance.")

TM.done()

TM.display()

TM.save()
```

Exercise 4: Shopping Cart

Create a Product class with attributes name, price, and quantity. Implement a ShoppingCart class
that allows adding/removing products and calculating the total price. Store the cart contents in a
dictionary.

```python
#Define class product
class Product:
  def __init__(self, name, price, quantity):
    self.name = name
    self.price = price
    self.quantity = quantity

  def __str__(self):
    return f"Name {self.name}, Price {self.price}, Qty {self.quantity}"

#Define class ShoppingCart with multiple functions
class ShoppingCart:
  def __init__(self):
    self.cart = {}

  def add(self, name, price, quantity):
    self.cart[name] = Product(name, price, quantity)
    print(f"Product {name} added successfully")

  def remove(self, name):
    if name in self.cart:
      del self.cart[name]
      print(f"\nProduct {name} removed successfully")
    else:
      print(f"\nProduct {name} not found")

  def total(self):
    total = sum(product.price * product.quantity for product in self.cart.values())
    return total

  def view(self):
    if not self.cart:
      print("Cart is empty")
    else:
      print("\nCart Contents:")
      for product in self.cart.values():
        print(f"Name: {product.name}, Price: {product.price}, Qty: {product.quantity}")


myCart = ShoppingCart()

myCart.add("Wireless Mouse", 25, 3)
myCart.add("Bluetooth Headphones", 50, 2)
myCart.add("USB-C Charger", 15, 5)
myCart.add("External Hard Drive", 80, 1)

myCart.view()

myCart.remove("Bluetooth Headphones")

myCart.view()

print(f"\nTotal Amout: {myCart.total()} $")
```

Exercise 5: Movie Database

Create a Movie class with attributes title, genre, and rating. Implement a MovieDatabase class that
can add movies, search by genre, and display all movies sorted by rating. Use a set to store unique
genres.

```python
#Define class Movie
class Movie:
  def __init__(self, title, genre, rating):
    self.title = title
    self.genre = genre
    self.rating = rating

  def __str__(self):
    return f"Title: {self.title}, Genre: {self.genre}, Rating: {self.rating}"

#Define class MovieDatabase with different functions
class MovieDatabase:
  def __init__(self):
    self.movies = []
    self.genres = set()

  def add(self, title, genre, rating):
    movie = Movie(title, genre, rating)
    self.movies.append(movie)
    self.genres.add(genre)

  def search(self, genre):
    result = [movie for movie in self.movies if movie.genre.lower() == genre.lower()]
    if result:
      print(f"\nSearch by genre:")
      for movie in result:
        print(movie)
    else:
      print(f"No movies found in genre {genre}")

  def displySorted(self):
    sorted_movies = sorted(self.movies, key=lambda movie: movie.rating, reverse=True)
    print("\nAll movies sorted by rating:")
    for movie in sorted_movies:
      print(movie)

  def displayAll(self):
    print("All movies in database: ")
    for movie in self.movies:
      print(movie)


myMovie = MovieDatabase()

myMovie.add("The Matrix", "Sci-Fi", 8.7)
myMovie.add("Forrest Gump","Drama", 8.8)
myMovie.add("The Lion King", "Animation", 8.5)
myMovie.add("The Dark Knight", "Action", 9.0)

myMovie.displayAll()

myMovie.search("Sci-Fi")

myMovie.displySorted()
```

Exercise 6: Phonebook

Create a Contact class with attributes name, phone, and email. Implement a Phonebook class that allows adding, removing, and searching for contacts. Store contacts in a dictionary, and save/load the phonebook from a text file.

```python
#Define class contact with attributes
class Contact:
  def __init__(self, name, phone, email):
    self.name = name
    self.phone = phone
    self.email = email

  def __str__(self):
    return f"Name: {self.name}, Phone: {self.phone}, Email: {self.email}"

#Define class Phonebook with multiple functions
class Phonebook:
  def __init__(self):
    self.contacts = {}

  def add(self, name, phone, email):
    contact = Contact(name, phone, email)
    self.contacts[phone] = contact
    print(f"Contact {name} added successfully")

  def display(self):
    print("\nAll contacts in phonebook: ")
    for contact in self.contacts.values():
      print(contact)

  def remove(self, phone):
    if phone in self.contacts:
      del self.contacts[phone]
      print(f"Contact removed successfully")
    else:
      print(f"Contact not found")

  def search(self, name):
    print(f"\nSearch by name:")
    results = [contact for contact in self.contacts.values() if name.lower() in contact.name.lower()]
    if results:
      for contact in results:
        print(f"Name: {contact.name}, Phone: {contact.phone}, Email: {contact.email}")
    else:
      print(f"No contacts found for {name}")

  def save(self, filename = "phonebook.txt"):
    with open(filename, "w") as file:
      for contact in self.contacts.values():
        file.write(f"{contact.name},{contact.phone},{contact.email}\n")

  def load(self, filename = "phonebook.txt"):
    print("\nCurrent phonebook")
    with open(filename, "r") as file:
      for line in file:
        parts = line.strip().split(",")
        if len(parts) == 3:
          name, phone, email = parts
          contact = Contact(name, phone, email)
          print(contact)


phonebook = Phonebook()

phonebook.add("John Doe", "1234567890", "john.mclean@example.com")
phonebook.add("Jane Smith", "9876543210", "james.wayne@example.com")
phonebook.add("Alice", "1245789630", "alice@gmail.com")

phonebook.display()

phonebook.remove("1234567890")

phonebook.display()

phonebook.search("Alice")

phonebook.save()

phonebook.load()
```

Exercise 7: Stack Implementation

Implement a Stack class with methods to push, pop, and check if the stack is empty. Write a
function that checks if a given string of parentheses is balanced using your Stack class.

```python
#Implement class Stack
class Stack:
  def __init__(self):
    self.stack = []

  def push(self, element):
    self.stack.append(element)

  def pop(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack.pop()

  def peek(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack[-1]

  def isEmpty(self):
    return len(self.stack) == 0

  def size(self):
    return len(self.stack)


#Create a stack
myStack = Stack()

#Push item into stack
myStack.push("A")
myStack.push("B")
myStack.push("C")

print("Stack: ", myStack.stack)

#Pop item from stack
print("Pop: ", myStack.pop())
print("Stack: ", myStack.stack)

#Peek item in stack
print("Peek: ", myStack.peek())

#Check if stack is empty
print("Is empty: ", myStack.isEmpty())

#Get size of stack
print("Size: ", myStack.size())
```

Exercise 8: Queue Implementation

Implement a Queue class with methods to enqueue, dequeue, and check if the queue is empty.
Create a function that simulates a ticketing system where customers join the queue and are served
in order.

```python
#Implement Queue class
class Queue:
  def __init__(self):
    self.queue = []

  def enqueue(self, element):
    self.queue.append(element)

  def dequeue(self):
    if self.isEmpty():
      return "Queue is empty"
    return self.queue.pop(0)

  def peek(self):
    if self.isEmpty():
      return "Queue is empty"
    return self.queue[0]

  def isEmpty(self):
    return len(self.queue) == 0

  def size(self):
    return len(self.queue)


def ticketing_system(customers):
  myQueue = Queue()

  print("Enqueue: ")
  for customer in customers:
    myQueue.enqueue(customer)
    print(f"{customer} joined the queue")

  print("\nTicketing System: ", myQueue.queue)

  print("\nIs empty queue? : ", myQueue.isEmpty())

  print("\nDequeue: ")
  while not myQueue.isEmpty():
    customer = myQueue.dequeue()
    print(f"{customer} is served")


customers = ["John", "Alice", "Bob"]

ticketing_system(customers)
```

Exercise 9: Recipe Book

Create a Recipe class with attributes name, ingredients, and instructions. Implement a RecipeBook
class that allows adding, removing, and searching for recipes. Store recipes in a JSON file.

```python
import json

#Implement class Recipe with multiple attributes
class Recipe:
  def __init__(self, name, ingredients, instructions):
    self.name = name
    self.ingredients = ingredients
    self.instructions = instructions

#Implement RecipeBook class with different functions
class RecipeBook:
  def __init__(self):
    self.recipes = []

  def add(self, name, ingredients, instructions):
    recipe = Recipe(name, ingredients, instructions)
    self.recipes.append(recipe)

  def remove(self, name):
    self.recipes = [recipe for recipe in self.recipes if recipe.name != name]

  def search(self, name):
    results = [recipe for recipe in self.recipes if name.lower() in recipe.name.lower()]
    if results:
      for recipe in results:
        print(f"Name: {recipe.name} Ingredients: {recipe.ingredients} Instructions: {recipe.instructions}")
    else:
      print("No matching recipes found")

  def display(self):
    for recipe in self.recipes:
      print(f"Name: {recipe.name}, Ingredients: {recipe.ingredients}, Instructions: {recipe.instructions}")

  def save(self, filename = "recipes.json"):
    data = []

    for recipe in self.recipes:
       data.append({
        "name": recipe.name,
        "ingredients" : recipe.ingredients,
        "instructions" : recipe.instructions
      })

    with open(filename, "w") as file:
      json.dump(data, file, indent=4)

  def load(self, filename):
    with open(filename, "r") as file:
      data = json.load(file)

    print(json.dumps(data, indent=4))

RB = RecipeBook()

RB.add("Recipe1", ["ingredient1", "ingredient2"], "instruction1")
RB.add("Recipe2", ["ingredient3", "ingredient4"], "instruction2")
RB.add("Recipe3", ["ingredient5", "ingredient6"], "instruction3")

print("Initial recipe book \n")
RB.display()

print("\nSearching by Recipe1")
RB.search("Recipe1")

RB.remove("Recipe1")

print("\nCurrent recipe book \n")
RB.display()

RB.save()

print("\nLoad the JSON data\n")
RB.load("recipes.json")
```

Exercise 10: Weather Data Analyzer

Create a WeatherData class to store temperature data for a week. Implement methods to calculate
the average temperature, highest, and lowest temperatures. Store data in a tuple and read/write
from/to a CSV file.

```python
#Define class WeatherData with member of tuple
class WeatherData:
  def __init__(self, *temperatures):
    self.temperatures = temperatures

  #Add data to the tuple
  def add(self, data):
    self.temperatures.append(self.data)

  def __str__(self):
    return f"{self.temperatures}"

  #Calculate the average
  def average(self):
    return sum(self.temperatures) / len(self.temperatures)

  #Find the max value
  def highest(self):
    return max(self.temperatures)

  #Find the min value
  def lowest(self):
    return min(self.temperatures)

  #Save data into file
  def save(self, filename = "temperature.csv"):
    with open(filename, "w") as file:
      for temperature in self.temperatures:
        file.write(f"{temperature} \n")

  #Read data from file
  def read(self, filename = "temperature.csv"):
    print("\nTemperature data: ")
    with open(filename, "r") as file:
      for data in file:
        temperature = data.strip()
        print(temperature)


WD = WeatherData(10, 20, 30, 40, 50)

print("\nTemperature data for week: ", WD)

print("\nAverage temperature: ", WD.average())

print("\nHighest temperature: ", WD.highest())

print("\nLowest temperature: ", WD.lowest())

WD.save()

WD.read()
```
